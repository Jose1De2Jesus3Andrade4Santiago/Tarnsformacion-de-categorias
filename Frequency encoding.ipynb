# Frequency Encoding

```python
import pandas as pd
```

```python
#Cargamos el DataFrame
df = pd.read_csv('ganaderia2.csv')
df.head()
```

```python
# Extaemos la columna del Data Frame
column = df.loc[:,'Producto']
column.head()
column.value_counts()
```

```python
# Ponemos la columna en una lista
column = column.values
column = column.tolist()
```

```python
# Creamos un arreglo aux igual a la columna del Data Frame
aux =[]
aux= column[:]
```

```python
for i in range(len(aux)):
    aux[i] = column.count(aux[i])
```

```python
# Convertimos la lista en DataFrame
aux = pd.DataFrame(aux) 
```

```python
#Añadimos aux al DataFrame original 
df.insert(8,'Producto_Fe',aux)
```
